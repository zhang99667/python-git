# Numpy 常用 random 随机函数汇总

官方文档地址：https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html

```python
import numpy as np
np.random.seed(666)
```

## 1.rand(d0, d1, ..., dn)

返回数据在 `[0, 1)` 之间，具有 **均匀分布**

```python
np.random.rand(5) # 一维数组，包含五个元素
```

```python
np.random.rand(3, 4) # 三行四列随机数
```

```python
np.random.rand(2, 3, 4) # 两块，每块儿三行四列
```

## 2.randn(d0, d1, ..., dn)
返回数据具有 **标准正态** 分布（均值0，方差1）

```python
np.random.randn(5)
```

```python
np.random.randn(3, 4)
```

```python
np.random.randn(2, 3, 4)
```

## 3.randint(low[, high, size, dtype])

生成随机整数，包括 low，不包括 high

如果 high 不指定，则从 `[0,low)` 中生成数字

```python
np.random.randint(3) # [0,3) 中的数字
```

```python
np.random.randint(1, 10) #[1, 10) 中的数字
```

```python
np.random.randint(10, 30, size=(5,)) # 用元组定义维度。 一维数组五个元素，区间 [10,30)
```

```python
np.random.randint(10, 30, size=(2, 3, 4)) # 两块，三行四列，区间 [10,30)
```

## 4.random([size])

生成 `[0.0, 1.0]` 的随机数

```python
np.random.random(5)
```

```python
np.random.random(size=(3, 4)) # 三行四列
```

```python
np.random.random(size=(2, 3, 4)) # 两块三行四列
```

## 5.choice(a[, size, replace, p])

a 是一维数组，从它里面生成随机结果

```python
# 这时候，a 是数字，则从 range(5) 中生成，size 为 3
np.random.choice(5, 3) # 从 range(5) 中随机抽取 3 个数字
```

```python
np.random.choice(5, (2, 3)) # 从 range(5) 中随机抽取能组成 两行三列 的数字
```

```python
array = [2, 3, 6, 7, 9]
np.random.choice(array, 3) # 从数组里随机抽取三个数字
```

```python
np.random.choice(array, (2, 3)) # 从数组里随机抽取能组成 两行三列 的数字
```

## 6.shuffle(x)

把一个数组 x 进行打散后随机排列

```python
a = np.arange(10)
np.random.shuffle(a) # 打散后随机排列，组成一行
a
```

```python
a = np.arange(20).reshape(4, 5) # 打散后随机排列，组成四行五列
a
```

```python
# 如果数组是多维的，则只会在第一维度打散数据
# 只打散行之间的数据，列之间的数据不打散
np.random.shuffle(a)
a
```

## 7.permutation(x)

把一个数组 x 进行随机排列，或者数字的全排列

```python
# 这时候，生成 range(10) 的随机排列
np.random.permutation(10)
```

```python
# 生成一个 3×3 数组
arr = np.arange(9).reshape(3, 3)
arr
```

```python
# 这时候，在第一维度进行打散
np.random.permutation(arr)
```

```python
arr # 注意，这里不会改变原来得 arr，会返回一个新的 copy
```

## 8.normal([loc, scale, size])

按照平均值 loc 和方差 scale 生成高斯分布的数字

```python
np.random.normal(1, 10, 10)
```

```python
np.random.normal(1, 10, (3, 4))
```

## 9.uniform([low, high, size])

在 `[low, high)` 之间生成均匀分布的数字

```python
np.random.uniform(1, 10, 10)
```

```python
np.random.uniform(1, 10, (3, 4))
```

## 实例：对数组加入随机噪声

```python
import matplotlib.pyplot as plt
```

```python
# 绘制 sin 曲线
x = np.linspace(-10, 10, 100) # linspace(start, end, size)函数常用的有三个参数：start：起点、stop：终点、num: 生成数据的数量；
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# 加入噪声
x = np.linspace(-10, 10, 100)
y = np.sin(x) + np.random.rand(len(x)) # 生成均匀分布的噪声
plt.plot(x, y)
plt.show()
```
